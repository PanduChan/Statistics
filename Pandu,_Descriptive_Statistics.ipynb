# <b>C. <span style='color:#3A6D8C'>Central Tendency</span></b>

Ukuran pemusatan adalah ukuran statistik yang digunakan untuk menggambarkan pusat atau titik-titik sentral dari sebuah distribusi data. Ukuran pemusatan memberikan informasi tentang nilai "*typical*" atau representatif dari kumpulan data.

Ukuran pemusatan yang umum digunakan adalah :
* Mean (rata-rata)
* Median (nilai tengah)
* Modus (nilai yang paling sering muncul)
* Kuartil (membagi data menjadi 4 bagian)

Pemilihan ukuran pemusatan yang tepat tergantung pada sifat data dan tujuan analisis statistik yang dilakukan.

## <b>C.1. <span style='color:#3A6D8C'> Mean ( *XÌ„* )</span></b>
---

Mean adalah rata-rata dari angka-angka yang diberikan dan dihitung dengan membagi jumlah angka-angka tersebut dengan total jumlah angka.<br><br>

\begin{equation}
\bar{X} =  \frac{Jumlah\;Data}{Banyak\;Data}
\end{equation}

\begin{cases}

# <b>D. <span style='color:#3A6D8C'>Dispersion</span></b>

Dispersion (penyebaran) dalam statistika merujuk pada ukuran variabilitas atau sebaran data. Hal ini menggambarkan sejauh mana nilai-nilai data tersebar di sekitar pusat distribusi

Ukuran penyebaran yang umum digunakan adalah :
* Range (rentang)
* Simpangan Baku dan Varians
* Interquartile Range (Selisih kuartil 1 dan 3)

## <b>D.1. <span style='color:#3A6D8C'>Range</span></b>
---

Rentang (range) adalah ukuran statistik yang menggambarkan selisih antara nilai maksimum dan nilai minimum dalam suatu set data. Dalam konteks statistika, rentang memberikan gambaran kasar tentang sebaran data, yaitu sejauh mana nilai-nilai data tersebar di antara nilai maksimum dan minimum.<br><br>

\begin{equation}
Range =  Max\;(Data)- Min\;(Data)

## <b>D.2. <span style='color:#3A6D8C'>Standard Deviation & Variance</span></b>
---

Simpangan baku (standard deviation) adalah ukuran statistik yang digunakan untuk mengukur sebaran atau dispersi data dalam sebuah sampel atau populasi. Simpangan baku menggambarkan sejauh mana nilai-nilai data tersebar di sekitar rata-rata (mean) dalam suatu distribusi.

\begin{equation}
\sigma = \sqrt{\frac{\sum{(x - \mu)^2}}{N}}
\end{equation}

* Ïƒ adalah simpangan baku.
* ð’™ adalah nilai data dalam sampel atau populasi.
* Î¼ adalah rata-rata (mean) dari data.

<br><br>

\begin{equation}
var = \sigma^2

# <b>E. <span style='color:#3A6D8C'>Outliers</span></b>

Outliers (pencilan) adalah nilai-nilai yang sangat berbeda atau jauh dari nilai-nilai lain dalam suatu dataset. Mereka adalah observasi yang ekstrem atau tidak biasa dibandingkan dengan nilai-nilai lainnya. Outliers bisa menjadi hasil dari kesalahan pengukuran, perbedaan dalam populasi atau sebenarnya, atau mungkin merupakan indikasi dari variasi yang signifikan dalam data.

Outliers dapat diidentifikasi menggunakan metode statistik, seperti menggunakan batas atas dan batas bawah berdasarkan kriteria tertentu, atau dengan memvisualisasikan data menggunakan grafik seperti box plot atau scatter plot. Pendekatan yang umum digunakan adalah dengan menggunakan metode seperti kuartil (interquartile range/IQR) atau tiga kali deviasi standar.
<br><br>

\begin{equation}
\text{Lower Fence} = Q_1 - 1.5 * IQR\\
\text{Upper Fence} = Q_3 + 1.5 * IQR
\end{equation}
<br>

Jika sebuah nilai pada data dibawah nilai Lower Fence disebut *low outlier* sedangkan nilai data diatas Upper Fence disebut *high outlier* <br><br>

**Perhatikan Contoh Berikut :**

Pada Python, akan dibuat outlier labeling pada sebuah data yakni dengan kriteria

* Jika nilai dibawah Lower Hence maka beri label 'Low Outlier'
* Jika nilai diatas Upper Hence maka beri label 'High Outlier'
* Selain itu beri label 'InLier' (lawan kata outlier)

````python
# Proses penentuan ambang batas outlier

def hitung_batas_outlier(data):
    Q1 = data.quantile(q = 0.25)
    Q3 = data.quantile(q = 0.75)
    IQR = Q3 - Q1
    lower_hence = Q1 - 1.5 * IQR
    upper_hence = Q3 + 1.5 * IQR
    return(lower_hence, upper_hence)
````

````python
# Proses pemberian label outlier pada data

def beri_label_outlier(nilai, lower_fence, upper_fence):
    if(nilai < lower_fence):
        tipe = 'Low Outlier'
    elif(nilai > upper_fence):
        tipe = 'High Outlier'
    else:
        tipe = 'InLier'

    return(tipe)
````

````python
# Panggil fungsi untuk menghitung batas outlier
lower_fence, upper_fence = hitung_batas_outlier(data = gas_emissions['Emissions'])

# Tampilkan hasilnya
print(f'Lower Hence pada kolom Emissions ={lower_fence : .2f}')
print(f'Upper Hence pada kolom Emissions ={upper_fence : .2f}')
````

````python
# Setelah lower_hence dan upper_hence ditemukan,
gas_emissions['Type Outlier Emissions'] = gas_emissions['Emissions'].apply(beri_label_outlier, args = (lower_fence, upper_fence))

# Tampilkan hasilnya
display(gas_emissions[['Emissions', 'Type Outlier Emissions']])
````

# <b>G. <span style='color:#3A6D8C'>Skewness</span></b>

Kemiringan (skewness) adalah sebuah ukuran statistik yang menggambarkan asimetri atau ketidaksimetrisan dari distribusi data. Ini memberikan informasi tentang sejauh mana ekor distribusi data miring ke kiri atau ke kanan relatif terhadap pusat distribusi.

\begin{equation}
\text{Skewness} = \frac{\sum_{i=1}^{N}(x_i - \bar{x})^3}{N \cdot \sigma^3}
\end{equation}
<br>

Jika
* Skewness < 0 maka disebut Negatively Skewed
* Skewness = 0 maka disebut Normal
* Skewness > 0 maka Positively Skewed

# <b>H. <span style='color:#3A6D8C'>Korelasi</span></b>

Korelasi adalah sebuah konsep statistik yang mengukur sejauh mana dua variabel berkaitan satu sama lain. Korelasi menggambarkan hubungan linier antara dua variabel, yaitu sejauh mana perubahan dalam satu variabel berhubungan dengan perubahan dalam variabel lainnya. Dua variabel dikatakan berkorelasi apabila perubahan salah satu variabel disertai dengan perubahan variabel lainnya, baik dalam arah yang sama ataupun arah yang sebaliknya.

## **8.2 HeatMap**

Heatmap adalah visualisasi yang berguna untuk memvisualisasikan korelasi antara dua set data atau antara beberapa variabel dalam bentuk matriks.

docs : *https://plotly.com/python/heatmaps/*

````python
# Import library yang dibutuhkan
import plotly.express as px

# Hitung korelasi
korelasi = data_superstore[['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost']].corr()

# Buat Heat Map
fig = px.imshow(
    korelasi,
    text_auto = True,
    color_continuous_scale = 'Blues'
)

# Jangan tampilkan color axes / legend
fig.update_coloraxes(
    showscale = False
)

# Percantik tampilan
fig.update_layout(
    width = 800,
    height = 600,
    title = dict(
        text = "<b>Korelasi Antar Variable pada Data Super Store</b>",
        font = dict(
            size = 19,
            color = '#0E2954'
        ),
        y = 0.95,
        x = 0.5
    ),
    margin = dict(
        t = 80,
        b = 30,
        r = 50,
        l = 50
    )
)

# Tampilkan hasil visualisasi
fig.show()
````

````python
# Import library yang dibutuhkan
import plotly.express as px

# Buat scatter plot
fig = px.scatter(
    data_superstore,
    x = 'Sales',
    y = 'Shipping Cost',
    trendline = 'ols',
    color_discrete_sequence  = ['#0E2954']
)

# Tambahkan judul
fig.update_layout(
    title = dict(
        text = "<b>Korelasi Antar Variable Sales dan Shipping Cost<br>pada Data Super Store</b>",
        font = dict(
            size = 19,
            color = '#0E2954'
        ),
        y = 0.95,
        x = 0.5
    ),
)

# Tampilkan hasil visualisasinya
fig.show()
````
